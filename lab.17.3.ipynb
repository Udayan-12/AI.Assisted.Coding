```python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Generate sample data for ride-sharing
data_rideshare = {
    'ride_id': range(101, 121),
    'pickup_location': [
        '  Downtown ',
        'Airport',
        'suburbia',
        'downtown',
        'Uptown',
        'Airport ',
        'SUBURBIA',
        'Downtown',
        'airport',
        'Uptown',
        'Downtown',
        'Suburbia',
        'Airport',
        'Uptown',
        'Downtown',
        'Airport',
        'Suburbia',
        'Uptown',
        'Downtown',
        'Airport'
    ],
    'dropoff_location': [
        'Uptown',
        'Downtown',
        'Airport',
        'Uptown ',
        'Downtown',
        'suburbia',
        'Airport',
        'Uptown',
        'Downtown',
        'Airport',
        'Uptown',
        'Downtown',
        'Airport',
        'Uptown',
        'Downtown',
        'Airport',
        'Uptown',
        'Downtown',
        'Airport',
        'Uptown'
    ],
    'fare_amount': np.random.normal(25, 10, 20).round(2),
    'driver_rating': np.random.uniform(2.5, 5.0, 20).round(1),
    'trip_duration_minutes': np.random.randint(10, 60, 20)
}

df_rideshare = pd.DataFrame(data_rideshare)

# Introduce some missing values
missing_fare_indices = np.random.choice(df_rideshare.index, 3, replace=False)
df_rideshare.loc[missing_fare_indices, 'fare_amount'] = np.nan

# Introduce some duplicate entries (based on ride_id and other key features)
duplicate_rows = df_rideshare.sample(2, random_state=1).copy()
duplicate_rows['ride_id'] = [121, 122] # Assign new ride_ids for duplicates for realism
df_rideshare = pd.concat([df_rideshare, duplicate_rows], ignore_index=True)

# Introduce some inconsistent driver ratings (outside typical 1-5 scale or just unusual)
# Let's say one driver got a 0.5 and another a 5.2
df_rideshare.loc[np.random.choice(df_rideshare.index, 1), 'driver_rating'] = 0.5
df_rideshare.loc[np.random.choice(df_rideshare.index, 1), 'driver_rating'] = 5.2

print("Sample Ride-Sharing DataFrame created:")
display(df_rideshare.head(10))
print("\nDataFrame Information:")
df_rideshare.info()
```

```python
# 4. Normalize driver ratings using Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler_driver_rating = MinMaxScaler()

# Reshape the column for scaling (MinMaxScaler expects a 2D array)
df_rideshare['driver_rating_normalized'] = scaler_driver_rating.fit_transform(df_rideshare[['driver_rating']])

print("\nDataFrame after normalizing 'driver_rating':")
display(df_rideshare[['driver_rating', 'driver_rating_normalized']].head())
df_rideshare.info()
```

```python
# 3. Remove duplicate ride entries
initial_rows_rideshare = len(df_rideshare)
df_rideshare.drop_duplicates(subset=['ride_id', 'pickup_location', 'dropoff_location', 'trip_duration_minutes'], inplace=True)
duplicates_removed_rideshare = initial_rows_rideshare - len(df_rideshare)

if duplicates_removed_rideshare > 0:
    print(f"Removed {duplicates_removed_rideshare} duplicate ride entry(s).")
else:
    print("No duplicate ride entries found.")

print("\nDataFrame after removing duplicates:")
display(df_rideshare.head())
df_rideshare.info()
```

```python
# 2. Fill missing fare amounts with median values
median_fare = df_rideshare['fare_amount'].median()
df_rideshare['fare_amount'] = df_rideshare['fare_amount'].fillna(median_fare)

print(f"Missing 'fare_amount' values imputed with median: {median_fare}")
print("\nDataFrame after filling missing fare amounts:")
display(df_rideshare.head())
df_rideshare.info()
```
